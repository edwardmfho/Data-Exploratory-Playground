<h4>Getting Movie ID</h4>
<p>To fill in the missing data, we would need to first create a new column, and get the movie id for each movie.</p>

```python
import requests
from configs import TMDB_API
```

```python
# Find Movie ID in TMDB
def get_movie_id(movie_name):
    response = requests.get(
    'https://api.themoviedb.org/3/search/movie',
    params={'api_key': TMDB_API,
            'query' : movie_name},
    )
    
    try:
        movie_id = response.json()['results'][0]['id']     
#       print('Movie ID found: ', movie_id)
    except:
        movie_id = np.nan
      
    return movie_id
```

```python
movie_df['movie_id'] = np.nan
```

```python
movie_df[:10]
```

```python
movie_df['movie_id'] = movie_df.apply(lambda x: get_movie_id(x['title']), axis=1)
```

```python
def find_director(row):
    movie_id = int(row['movie_id'])   
    response = requests.get(
    'https://api.themoviedb.org/3/movie/{}/credits'.format(movie_id),
    params={'api_key': TMDB_API,},
    )

    director = ''
    for crew in response.json()['crew']:
        if crew['job'] == 'Director':
            director += crew['original_name'] + ', '
#         print(director)
    return str(director[:-1])


def find_cast(row):
    movie_id = int(row['movie_id'])
    response = requests.get(
    'https://api.themoviedb.org/3/movie/{}/credits'.format(movie_id),
    params={'api_key': TMDB_API,},
    )

    cast_list = ''
    for cast in response.json()['cast']:
        cast_list += cast['name'] + ', '
#     print(cast_list)
    return str(cast_list[:-1])


def find_country(row):
    movie_id = int(row['movie_id'])


#             print('finding country')
    response = requests.get(
    'https://api.themoviedb.org/3/movie/{}'.format(movie_id),
    params={'api_key': TMDB_API,},
    )

#     print(response.status_code)
    try:
        country_list = ''
        for country in response.json()['production_countries']:
            country_list += country['name'] + ', '
        
#         country_list = response.json()['production_countries'][0]['name']
#         print('country list: ', country_list)
        return str(country_list[:-1])
    except:
        return(row['country'])



```

```python
missing_data = movie_df[movie_df[['country','director','cast']].isnull().any(axis=1)]
```

```python
# drop missing data with no movie_id

missing_data = missing_data.dropna(subset=['movie_id']).copy()
```

```python
missing_data['country'] = missing_data.apply(lambda x: find_country(x), axis=1)
```

```python
missing_data['director'] = missing_data.apply(lambda x: find_director(x), axis=1)
```

<p>We could start by removing the "min" in the duration column, which could be done by a simple function.</p>

```python
movie_df.iloc[0]['duration']
```

```python
def remove_unit(input_value, unit_size):
    value = input_value[:-unit_size]
    return value
```

```python
movie_df['duration'] = movie_df.apply(lambda x: remove_unit(x['duration'], 4), axis=1)
```

```python
movie_df.head(3)
```

<p>Also we should check if the country names have been consistent, for example I noticed that USA was named "United States" in the original data, but "United States of America" in the TMDB data. So we should check if we need to conduct a bit of preprocessing there.</p>

```python
def replace_country_name(country_list, find_word, replaced_by):
    return country_list.replace(find_word, replaced_by).rstrip(',')
```

```python
movie_df['country'].value_counts()
```

```python
movie_df['country'] = movie_df.apply(lambda x: replace_country_name(x['country'], 'United States of America', 'United States'), axis=1)
```

```python
movie_df['country'].value_counts()
```

```python
country_df = movie_df['country'].str.get_dummies(', ')
```

```python
combined_movie_df = pd.concat([movie_df, country_df], axis=1)
```

```python
combined_movie_df.drop(columns = 'country', axis = 1, inplace = True)
```

```python
combined_movie_df.head(5)
```

<p>Same goes for the cast, listed_in and director columns. </p>

```python
director_df = movie_df['director'].str.get_dummies(', ')
cast_df = movie_df['cast'].str.get_dummies(', ')
genre_df = movie_df['listed_in'].str.get_dummies(', ')

combined_movie_df = pd.concat([combined_movie_df, director_df], axis=1)
combined_movie_df = pd.concat([combined_movie_df, cast_df], axis=1)
combined_movie_df = pd.concat([combined_movie_df, genre_df], axis=1)

combined_movie_df.drop(columns = ['director', 'cast', 'listed_in'])
```
