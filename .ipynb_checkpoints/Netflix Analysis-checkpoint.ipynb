<p>Missing data of movies should be relatively easy to resolve, here, I would make use of the <a href="https://www.themoviedb.org/">TMDB (The Movie Database) API</a> to fill in the missing data.</p>

<p>So I have create 4 simple functions to deal with this problem (please use your own API keys, it is free.).</p>

<h4>Getting Movie ID</h4>
<p>To fill in the missing data, we would need to first create a new column, and get the movie id for each movie.</p>

```python
import requests
```

```python
TMDB_API = '476e30bfe7878c3cf343c4446566b6ec'

# Find Movie ID in TMDB
def get_movie_id(movie_name):
    response = requests.get(
    'https://api.themoviedb.org/3/search/movie',
    params={'api_key': TMDB_API,
            'query' : movie_name},
    )
    
    try:
        movie_id = response.json()['results'][0]['id']     
#       print('Movie ID found: ', movie_id)
    except:
        movie_id = np.nan
      
    return movie_id
```

```python
movie_df['movie_id'] = np.nan
```

```python
movie_df[:10]
```

```python
movie_df['movie_id'] = movie_df.apply(lambda x: get_movie_id(x['title']), axis=1)
```

```python
movie_df
```

<p>Now we have filled up most <b>movie_id</b>, we should be able to easily retrieve the other data.</p>
<p>Before that, we should also save the file as a new checkpoint</p>

```python
pd.to_csv('movie_df.csv', index = False)
```

```python
def find_director(movie_id):
    response = requests.get(
    'https://api.themoviedb.org/3/movie/{}/credits'.format(movie_id),
    params={'api_key': TMDB_API,},
    )
    
    director = ''
    for crew in response.json()['crew']:
        if crew['job'] == 'Director':
            director += crew['original_name'] + ', '

    return director

def find_cast(movie_id):
    response = requests.get(
    'https://api.themoviedb.org/3/movie/{}/credits'.format(movie_id),
    params={'api_key': TMDB_API,},
    )

    cast_list = ''
    for cast in response.json()['cast']:
        cast_list += cast['name'] + ', '

    return cast_list

def find_country(movie_id):
    response = requests.get(
    'https://api.themoviedb.org/3/movie/{}'.format(movie_id),
    params={'api_key': TMDB_API,},
    )
    
    country_list = ''   
    for country in response.json()['production_companies']:
        country_list += country['origin_country'] + ', '

    return country_list

def fill_missing_data(data, column_type):
    
    movie_id = data['movie_id']
    movie_cast = data['cast']
    movie_director = data['director']
    movie_country = data['country']
        
    
    if movie_id != np.nan:
        if column_type == 'director' and movie_director == np.nan :
            result = find_director(int(movie_id))
        elif column_type == 'cast' and movie_cast == np.nan:
            result = find_cast(int(movie_id))
        elif column_type == 'country' and movie_country == np.nan:
            result = find_country(int(movie_id)) 
        else:
            pass
    else:
        pass
    return result
```

```python
movie_df['country'] = movie_df[10:].apply(lambda x: fill_missing_data(x, 'country'), axis=1)
movie_df['cast'] = movie_df[10:].apply(lambda x: fill_missing_data(x, 'cast'), axis=1)
movie_df['director'] = movie_df[10:].apply(lambda x: fill_missing_data(x, 'director'), axis=1)
```
